```python
# Do it with a Function

def fetch_weather_data(cities, api_key):
    
    weather_data = []

    for city in cities:
        response = requests.get(f"http://api.openweathermap.org/data/2.5/forecast?q={city}&appid={API_key}&units=metric")
        if response.status_code == 200:
            data = response.json()
            lat = data['city']['coord']['lat']
            lon = data['city']['coord']['lon']
            population = data['city'].get('population', 'Not available')

            first_forecast = data['list'][0]
            temp = first_forecast['main']['temp']
            temp_min = first_forecast['main']['temp_min']
            temp_max = first_forecast['main']['temp_max']
            weather_description = first_forecast['weather'][0]['description']
            wind_speed = first_forecast['wind']['speed']
            humidity = first_forecast['main']['humidity']
            visibility = first_forecast.get('visibility', 'Not available')

            # Extracting rain data
            rain = first_forecast.get('rain', {}).get('3h', 0)  

            # Extracting snow data
            snow = first_forecast.get('snow', {}).get('3h', 0) 
            weather_data.append({
            "city": city, 
            "latitude": lat, 
            "longitude": lon, 
            "population": population,
            "temperature": temp,
            "temp_min": temp_min,
            "temp_max": temp_max,
            "weather_description": weather_description,
            "wind_speed": wind_speed, 
            "humidity": humidity, 
            "visibility": visibility,
            "rain_3h": rain,
            "snow_3h": snow
             })
        else:
            print(f"Failed to get data for {city}")

    return pd.DataFrame(weather_data)

# Input data 
API_key="4c6096516c608a8e259fc8fa013c8266"
cities = ["Berlin", "Frankfurt", "Cologne", "Munich", "Hamburg", "Leipzig", "Düsseldorf", "Stuttgart", "Dortmund", "Essen", "Bremen", "Dresden", "Hanover", "Nuremberg", "Bonn", "Münster"]
weather_df = fetch_weather_data(cities, API_key)
weather_df['population'] = pd.to_numeric(weather_df['population'], downcast='float')
print(weather_df)

```
